```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv('INV_PROJECTION.csv', sep = ';', dtype = str)

list_reporting_week = data['Reporting week'].unique().tolist()
list_location = data['Location'].unique().tolist()
list_brand = data['Brand'].unique().tolist()
list_SNP_planner = data['SNP Planner'].unique().tolist()
list_technology = data['Technology'].unique().tolist()
list_production_line = data['Production Line'].unique().tolist()
list_key_figure = data['Key figure'].unique().tolist()
```

```python
data_week6 = data[data['Reporting week'] == '2018 w06']
data_week6_AE01 = data_week6[data_week6['Location'] == 'AE01']
data_week6_AE01_CNC = data_week6_AE01[data_week6_AE01['Brand'] == 'Clean & Clear']
data_week6_AE01_CNC_SNP = data_week6_AE01_CNC[data_week6_AE01_CNC['SNP Planner'] == 'P14']
data_week6_AE01_CNC_SNP_Technology = data_week6_AE01_CNC_SNP[data_week6_AE01_CNC_SNP['Technology'] == 'Mass Tubes T10 T18']
data_week6_AE01_CNC_SNP_prodline = data_week6_AE01_CNC_SNP[data_week6_AE01_CNC_SNP['Production Line'] == 'T18']


data_week6_AE01_CNC_SNP_prodline = data_week6_AE01_CNC_SNP[data_week6_AE01_CNC_SNP['Production Line'] == 'T18']
data_week6_AE01_CNC_SNP_prodline.set_index(['Reporting week', 'Location', 'Brand', 'SNP Planner', 'Technology', 'Production Line', 'Key figure'], inplace = True)
data_week6_AE01_CNC_SNP_prodline.index

```

```python
for i in range(0, data_week6_AE01_CNC_SNP_prodline.shape[1]) : 
    data_week6_AE01_CNC_SNP_prodline.iloc[:, i] = data_week6_AE01_CNC_SNP_prodline.iloc[:, i].str.replace(' ', '')
    data_week6_AE01_CNC_SNP_prodline.iloc[:,i] = pd.to_numeric(data_week6_AE01_CNC_SNP_prodline.iloc[:,i])
    
data_week6_AE01_CNC_SNP_prodline.info()
```

```python
data_week6_AE01_CNC_SNP_prodline_pivoted = data_week6_AE01_CNC_SNP_prodline.pivot_table(
    index= ['Reporting week', 'Location' , 'Brand', 'SNP Planner', 'Technology', 'Production Line'],
    columns = 'Key figure',
    values = data_week6_AE01_CNC_SNP_prodline.columns, 
    aggfunc= 'sum'
    )
data_week6_AE01_CNC_SNP_prodline_pivoted
```

```python
""" Maintenant que nous avons réussi pour une seule catégorie, nous automatisons pour le reste de l'Excel"""
"""1ère automatisation avec le reporting de la week 6 d'abord"""

data_week6 = data[data['Reporting week'] == '2018 w06']
data_week6.head()
data_week6.set_index(['Reporting week', 'Location', 'Brand', 'SNP Planner', 'Technology', 'Production Line', 'Key figure'], inplace = True)

data_week6.info()
```

```python
for i in range(0, data_week6.shape[1]) : 
    data_week6.iloc[:, i] = data_week6.iloc[:, i].str.replace(' ', '')
    data_week6.iloc[:,i] = pd.to_numeric(data_week6.iloc[:,i])
    data_week6.iloc[:,i] = data_week6.fillna(0)
    
data_week6.info()
```

```python
data_week6_pivoted = data_week6.pivot_table(
    index= ['Reporting week', 'Location' , 'Brand', 'SNP Planner', 'Technology', 'Production Line'],
    columns = 'Key figure',
    values = data_week6.columns, 
    aggfunc= 'sum'
    )
data_week6_pivoted
```

```python
data_week6_pivoted_stacked = data_week6_pivoted.stack(0)
data_week6_pivoted_stacked.to_csv('Week 6 Reporting.csv')
data_week6_pivoted_stacked.to_excel('Week 6 Reporting.xlsx')
```
